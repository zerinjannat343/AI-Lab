```python
data = pd.read_csv("student_kmeans_dataset.csv")
print(data.values)
```

```python
import numpy as np

dataset = np.array(data.values)
print(dataset)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

k = 3
kmeans = KMeans(n_clusters = k , random_state=42)
kmeans.fit(dataset)
center=kmeans.cluster_centers_
label=kmeans.labels_

plt.scatter(dataset[:, 0], dataset[:, 1], c=label)
plt.scatter(center[:, 0], center[:, 1], marker="*", s=300,color="red")
plt.title("Study Hours VS Attendance Percentage")
plt.xlabel("Study Hours per Week")
plt.ylabel("Attendance Percentage")
plt.grid(True)
```

```python
plt.scatter(dataset[:, 1], dataset[:, 2], c=label)
plt.scatter(center[:, 1], center[:, 2], marker="*", s=300, color="red")
plt.title("Attendance Percentage VS Avg Assignment Score")
plt.xlabel("Attendance Percentage")
plt.ylabel("Avg Assignment Score")
plt.grid(True)
```

```python
plt.scatter(dataset[:, 0], dataset[:, 2], c=label)
plt.scatter(center[:, 0], center[:, 2], marker="*", s=300, color="red")
plt.title("Study Hours VS Avg Assignment Score")
plt.xlabel("Study Hours ")
plt.ylabel("Avg Assignment Score")
plt.grid(True)
```
